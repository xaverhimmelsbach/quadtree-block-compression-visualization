```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Apply the default theme
sns.set_theme()

df = pd.read_csv('./measurements.csv', sep='\t')
df['compression_ratio'] = pd.to_numeric(
    df['compression_ratio'].str.replace(',', '.'))

df_sorted = df.sort_values('compression_ratio')

# Create a visualization
g = sns.relplot(
    x='file_name',
    y='compression_ratio',
    # yscale='log',
    data=df_sorted,
    height=5,
    aspect=16/9
)
g.set(xlabel='File name', ylabel='Compression ratio')

plt.xticks(rotation=45)

plt.show()
```
