```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Apply the default theme
sns.set_theme()
sns.set_style('whitegrid')
sns.set_context('paper')
```

```python
# Load data
df = pd.read_csv('../data/jpeg.csv')
#df['compression_ratio'] = pd.to_numeric(
#    df['compression_ratio'].str.replace(',', '.'))
```

```python
# Compression ratio
df_sorted = df.sort_values('compression_ratio')

# Create a visualization
g = sns.relplot(
    x='file_name',
    y='compression_ratio',
    # yscale='log',
    data=df_sorted,
    # height=5,
    aspect=16/9
)

# Draw line through 10‚Å∞
g.ax.axhline(y=10**0, color='k', linestyle='--', linewidth=1)

ticks = [0.1, 1, 10]
ticklabels = ['$10^{-1}$', '$10^0$', '$10^1$']
# ticklabels[1].set_fontweight('bold')
g.set(xlabel='File name', ylabel='Compression ratio', yscale='log', yticks=ticks, yticklabels=ticklabels)

plt.xticks(rotation=45)

plt.show()
```

```python
# Transcoding time
df_sorted = df.sort_values('transcoding_time')

# Create a visualization
g = sns.relplot(
    x='file_name',
    y='transcoding_time',
    # yscale='log',
    data=df_sorted,
    # height=5,
    aspect=16/9
)

ticks = [10, 100, 1000]
ticklabels = ['$10^1$', '$10^2$', '$10^3$']
g.set(xlabel='File name', ylabel='Transcoding time (ms)', yscale='log', yticks=ticks, yticklabels=ticklabels)

plt.xticks(rotation=45)

plt.show()
```
